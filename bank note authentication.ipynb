```python
#decision tree
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
#divide data in train and test
X = train_df.values[:,0:4]
Y = train_df.values[:,4]
X_train,X_test,y_train ,y_test =train_test_split(X,Y,test_size = 0.2)
```

```python
#Decision tree
C50_model = tree.DecisionTreeClassifier(criterion ='entropy').fit(X_train,y_train)


#predict new test cases
C50_predictions = C50_model.predict(X_test)
```

```python
#build confusion matrix
CM = pd.crosstab(y_test,C50_predictions)
#let us save TP,TN,FP,FN
TP=CM.iloc[1,1]
TN=CM.iloc[0,0]
FP=CM.iloc[0,1]
FN=CM.iloc[1,0]
```

```python
#check accuracy of the model
((TP+TN)*100)/(TP+TN+FP+FN)
```

```python
#false negative rate
((FN*100)/(FN+TP))
```

```python
#ROC CURVE
from sklearn import metrics
from sklearn.datasets import make_classification
from sklearn.metrics import roc_curve
from matplotlib import pyplot
fpr,tpr,threshold = metrics.roc_curve(y_test,C50_predictions)
plt.plot(fpr,tpr)
plt.xlim([0.0,1.1])
plt.ylim([0.0,1.1])
plt.title("ROC curve for decision tree")
plt.xlabel("False positive rate (1-specificity)")
plt.ylabel("True positive rate(sensitivity)")
plt.grid(True)
```

```python
#import libraries for logistic regression
from sklearn import tree
from sklearn.metrics import accuracy_score
```

```python
Sample_index = np.random.rand(len(train_df))<0.8

train = train_df[Sample_index]
test = train_df[~Sample_index]
```

```python
#select column index for independent variables 
train_cols = train.columns[0:4]
```

```python
#Build logistic regression 
import statsmodels.api as sm

logit = sm.Logit(train['e'],train[train_cols]).fit()
logit.summary()
```

```python
#predict test data
test['Actual_prob'] = logit.predict(test[train_cols])

test['Actual_val'] = 1
test.loc[test.Actual_prob <0.5,"Actual_val"] = 0
```

```python
#Build confusion matrix
CM =pd.crosstab(test['e'],test['Actual_val'])

#let us save TP.TN,FP,FN
TP=CM.iloc[1,1]
TN=CM.iloc[0,0]
FP=CM.iloc[0,1]
FN=CM.iloc[1,0]
```

```python
#accuracy of the model
((TP+TN)*100)/(TP+TN+FP+FN)
```

```python
#FALSE NEGATIVE RATE
(FN*100)/(FN+TP)
```

```python
#ROC CURVE
from sklearn import metrics
from sklearn.datasets import make_classification
from sklearn.metrics import roc_curve
from matplotlib import pyplot
fpr,tpr,threshold = metrics.roc_curve(test["e"],test["Actual_val"])
plt.plot(fpr,tpr)
plt.xlim([0.0,1.1])
plt.ylim([0.0,1.1])
plt.title("ROC curve for logistic regression")
plt.xlabel("False positive rate (1-specificity)")
plt.ylabel("True positive rate(sensitivity)")
plt.grid(True)
```
